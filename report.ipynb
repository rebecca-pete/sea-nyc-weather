### 1. Read and import data from the CSV files.

First, import the pandas, numpy, and matplotlib.pyplot libraries. 

The pandas library reads and imports data from the CSV files. The numpy and matplotlib.py libraries manipulate data and create visualizations, respectively.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### 2. Save the imported data&mdash;now an object or dataframe instance&mdash;into a variable.

Next, use the .read_csv() method to read the dataset you want to import and assign it to a variable. Run the following command to create a dataframe instance of Seattle weather data, and assign it to a variable labeled `SEA_data`. The resulting dataframe is an object, similar to a matrix. 

```python
SEA_data = pd.read_csv('data/KSEA.csv')
```

Use the following commands to convert `SEA_data.dtypes` and `NYC_data.dtypes` into dataframes, starting with Seattle.

```python
pd.DataFrame(SEA_data.dtypes, columns=['SEA DataTypes'])
```

```python
pd.DataFrame(NYC_data.dtypes, columns=['NYC DataTypes'])
```

Next, select the specific columns you want to use from each dataframe. Extract `date`, `actual_mean_temp`, `record_max_temp`, `record_max_temp_year`, `actual_precipitation`, and `record_precipitation`. 

Create two objects called `SEA_set` and `NYC_set`, each comprised of these six columns.

```python
SEA_set = SEA_data[['date', 'actual_mean_temp', 'record_max_temp', 'record_max_temp_year','actual_precipitation', 'record_precipitation']]
print(SEA_set)
```

```python
NYC_set = NYC_data[['date', 'actual_mean_temp', 'record_max_temp', 'record_max_temp_year', 'actual_precipitation', 'record_precipitation']]
print(NYC_set)
```

Next, rename those columns on the smaller datasets, `SEA_set` and `NYC_set`.

```python
SEA_set.rename(columns={'date': 'Date', 'actual_mean_temp': 'ActualMeanTemp', 'record_max_temp':'RecordMaxTemp', 'record_max_temp_year': 'RecordMaxTempYear','actual_precipitation':'ActualPrecip', 'record_precipitation':'RecordPrecip'}, inplace=True)
print(SEA_set)
```

```python
NYC_set.rename(columns={'date': 'Date', 'actual_mean_temp': 'ActualMeanTemp', 'record_max_temp':'RecordMaxTemp', 'record_max_temp_year': 'RecordMaxTempYear', 'actual_precipitation':'ActualPrecip', 'record_precipitation':'RecordPrecip'}, inplace=True)
print(NYC_set)
```

### 4. Merge the filtered dataframe instances together.

Next, join the two smaller dataframes or sets by the date.

```python
full_set=pd.merge(SEA_set, NYC_set, on='Date', suffixes=('_SEA', '_NYC'))
full_set
```

Next, check for any null values in the new object. 

```python
print(full_set.shape)
full_set.isnull().sum()
full_set.dtypes
```

There are no null datatypes in the object. The data is clean.

Before resetting the index to the date, create new columns showing the month and year.

```python
full_set['Date'] = pd.to_datetime(full_set['Date'])

full_set['Date_Year']  = full_set['Date'].dt.year
full_set['Date_Month'] = full_set['Date'].dt.month
full_set
```

Notice that the `Date_Month` and `Date_Year` columns have been added to the far right of the dataframe.

Reset the index to the primary key you want to use, in this case `Date`.

```python
full_set.set_index('Date', inplace=True)
full_set
```

### 5. Expose relationships between data points.

Next, check `ActualMeanTemp` for both cities over the twelve month period from July 2014 through June 2015.

```python
full_set_extract_mean_temp = full_set[['ActualMeanTemp_SEA', 'ActualMeanTemp_NYC']]
full_set_extract_mean_temp
```

### 6. Create data visualizations of your findings.

```python
full_set_rec_max_temp.plot(figsize=(15,5));
```

Next, compare `ActualPrecip_NYC` to `ActualMeanTemp_NYC`, this time using a scatter plot visualzation.

```python
full_set.plot(kind='scatter', x='ActualMeanTemp_NYC', y='ActualPrecip_NYC', figsize=(15,5));
```

New York City appears to have a higher precipitation over a wide temperature range. They have high precipitation from at least the 40-80 degree temperature rage. Next, see how that range compares to Seattle.

Run the following command to compare `ActualPrecip_SEA` to `ActualMeanTemp_SEA`.

```python
full_set.plot(kind='scatter', x='ActualMeanTemp_SEA', y='ActualPrecip_SEA', figsize=(15,5));
```

As you can see, Seattle seems to have a higher precipitation in the 45-65 degree range. This is a much smaller temperature range when compared to New York City.
